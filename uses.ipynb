```python
import IPython
import os
import soundfile as sf
from dotenv import load_dotenv
from huggingface_hub import login
from transformers.tools import HfAgent
load_dotenv()

def play_audio(audio):
    sf.write("speech_converted.wav", audio.numpy(), samplerate=16000)
    return IPython.display.Audio("speech_converted.wav")

login(os.getenv("HF_ACCESS_TOKEN"))
```

```python
agent_name = "StarCoder" #@param ["OpenAssistant", "StarCoder""]

agent_models = {
    "StarCoder" : "https://api-inference.huggingface.co/models/bigcode/starcoder",
}

open_assistant_agent_models = {
    "oasst-sft-4-pythia-12b-epoch-3.5" : "https://api-inference.huggingface.co/models/OpenAssistant/oasst-sft-4-pythia-12b-epoch-3.5",
    "pythia-12b-pre-v8-12.5k-steps" : "https://api-inference.huggingface.co/models/OpenAssistant/pythia-12b-pre-v8-12.5k-steps",
}

#agent = HfAgent(agent_models["StarCoder"]) #use with starcoder
agent = HfAgent(url_endpoint= open_assistant_agent_models['pythia-12b-pre-v8-12.5k-steps']) #use with open assistant

print("Model Activated")
```

```python
image = agent.run("generate a ultra realistic photo of spiderman dating capitan marvel in a spaceship")
image
```

```python
#write text /code
codes = agent.run("generate a html code for a login and register form on a website, MVC, ReactJS, TypeScript, arrow functions")
codes
```

```python
save_to_file = agent.run("save the codes in a a index.html file")
save_to_file
```

```python
# make music
audio = agent.run("make me a nice music")
play_audio(audio)
```
